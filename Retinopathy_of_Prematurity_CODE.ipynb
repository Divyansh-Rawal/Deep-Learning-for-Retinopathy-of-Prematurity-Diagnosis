```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.image as mpimg

# Paths to the confusion matrix images
image_path_v8 = '/content/ultralytics/runs/detect/train/confusion_matrix.png'
image_path_v11 = '/content/ultralytics/runs/detect/train3/confusion_matrix.png'

# Load the confusion matrix images
img1 = mpimg.imread(image_path_v8)
img2 = mpimg.imread(image_path_v11)

# Sample Confusion Matrices (Replace with actual values if available)
conf_matrix_v8 = np.array([[50, 10], [8, 120]])  # [[TN, FP], [FN, TP]]
conf_matrix_v11 = np.array([[55, 8], [6, 130]])  # [[TN, FP], [FN, TP]]

# Function to calculate accuracy, precision, recall, and F1 score
def calculate_metrics(conf_matrix):
    TN, FP, FN, TP = conf_matrix.ravel()
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    precision = TP / (TP + FP) if (TP + FP) != 0 else 0
    recall = TP / (TP + FN) if (TP + FN) != 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0
    return accuracy, precision, recall, f1_score

# Calculate metrics for YOLOv8
accuracy_v8, precision_v8, recall_v8, f1_v8 = calculate_metrics(conf_matrix_v8)

# Calculate metrics for YOLOv11
accuracy_v11, precision_v11, recall_v11, f1_v11 = calculate_metrics(conf_matrix_v11)

# Display Confusion Matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Show the first confusion matrix (YOLOv8)
sns.heatmap(conf_matrix_v8, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'], ax=axes[0])
axes[0].set_title('Confusion Matrix of YOLOv8')
axes[0].set_xlabel('Predicted Label')
axes[0].set_ylabel('True Label')

# Show the second confusion matrix (YOLOv11)
sns.heatmap(conf_matrix_v11, annot=True, fmt='d', cmap='Greens', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'], ax=axes[1])
axes[1].set_title('Confusion Matrix of YOLOv11')
axes[1].set_xlabel('Predicted Label')

# Display the heatmaps side by side
plt.tight_layout()
plt.show()

# Print the calculated metrics
print("=== YOLOv8 Model Performance ===")
print(f"Accuracy: {accuracy_v8:.4f}")
print(f"Precision: {precision_v8:.4f}")
print(f"Recall: {recall_v8:.4f}")
print(f"F1 Score: {f1_v8:.4f}")

print("\n=== YOLOv11 Model Performance ===")
print(f"Accuracy: {accuracy_v11:.4f}")
print(f"Precision: {precision_v11:.4f}")
print(f"Recall: {recall_v11:.4f}")
print(f"F1 Score: {f1_v11:.4f}")
```

```python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

# Simulated true labels and predictions (Replace with actual model outputs)
y_true = np.random.randint(0, 2, 100)  # Ground truth labels (0 or 1)
y_pred = np.random.randint(0, 2, 100)  # Model predictions (0 or 1)

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Plot confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Compute and display classification report
print(classification_report(y_true, y_pred, target_names=["Negative", "Positive"]))
```
